# ElasticNet Regression

```python
import numpy as np
import pandas as pd
from sklearn.linear_model import Ridge, Lasso, ElasticNet
from sklearn.metrics import mean_squared_error,r2_score
from sklearn.model_selection import train_test_split, cross_val_score
from sklearn import model_selection
import matplotlib.pyplot as plt
from sklearn.linear_model import RidgeCV, LassoCV, ElasticNetCV
```

```python
df = pd.read_csv("Hitters.csv")
df = df.dropna()
dms = pd.get_dummies(df[["League", "Division", "NewLeague"]]) # kategorig değişkenleri onehot encoding metoduyla dummy değişkenlere çevirdik
y = df["Salary"]
X_ = df.drop(["Salary", "League", "Division", "NewLeague"], axis=1).astype("float64")
X = pd.concat([X_, dms[["League_N", "Division_W", "NewLeague_N"]]], axis=1)
X_train, X_test, y_train, y_test = train_test_split(X,
                                                    y,
                                                    test_size=.25,
                                                    random_state=42)

df.head()
```

```python
enetModel = ElasticNet().fit(X_train, y_train)
```

```python
enetModel.intercept_
```

```python
enetModel.coef_
```

### Tahmin

```python
enetModel.predict(X_train)
```

```python
enetModel.predict(X_test)
```

```python
y_pred = enetModel.predict(X_test)
```

```python
RMSE = np.sqrt(mean_squared_error(y_test, y_pred)) 
RMSE
```

```python
r2_score(y_test, y_pred)
```

### Model Tuning

```python
alphas__ = pow(10, np.linspace(10, -2, 100))*.5
```

```python
enetCVModel = ElasticNetCV(alphas=alphas__, cv=10).fit(X_train, y_train)
```

```python
enetCVModel.alpha_
```

```python
enetCVModel.intercept_
```

```python
enetCVModel.coef_
```

### Final Model

```python
enetTuned = ElasticNet(alpha = enetCVModel.alpha_).fit(X_train, y_train)
```

```python
y_pred = enetTuned.predict(X_test)
```

```python
RMSE_Tuned=np.sqrt(mean_squared_error(y_test, y_pred))
RMSE_Tuned
```
